```python
import statsmodels.api as sm
from statsmodels.tsa.seasonal import STL

df = df.set_index("date")
df.index.freq = "MS"

stl = STL(df["attendances"], period=12)
result = stl.fit()
result.plot()
plt.show()
```

```python
train = df[df.index < "2024-01-01"]["attendances"]
test = df[df.index >= "2024-01-01"]["attendances"]

print(f"Training months: {len(train)}")
print(f"Test months: {len(test)}")
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX(
    train,
    order=(1, 1, 1),
    seasonal_order=(1, 1, 1, 12)
)

results = model.fit(disp=False)
print(results.summary())
```

```python
predictions = results.forecast(steps=len(test))
predictions.index = test.index

plt.figure(figsize=(14,5))
plt.plot(train.index, train, label="Training data")
plt.plot(test.index, test, label="Actual 2024-2026")
plt.plot(predictions.index, predictions, label="Forecast", linestyle="--")
plt.axvline(pd.Timestamp("2024-01-01"), color="black", linestyle=":", label="Forecast start")
plt.title("NHS A&E Attendances - SARIMA Forecast vs Actual")
plt.xlabel("Date")
plt.ylabel("Monthly Attendances")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(test, predictions)
rmse = np.sqrt(mean_squared_error(test, predictions))
mape = (abs(test - predictions) / test).mean() * 100

print(f"MAE:  {mae:,.0f} attendances")
print(f"RMSE: {rmse:,.0f} attendances")
print(f"MAPE: {mape:.1f}%")
```

```python
future_steps = 12

future_forecast = results.forecast(steps=len(test) + future_steps)
future_index = pd.date_range(
    start=test.index[0], 
    periods=len(test) + future_steps, 
    freq="MS"
)
future_forecast.index = future_index

plt.figure(figsize=(14,5))
plt.plot(train.index, train, label="Historical data")
plt.plot(test.index, test, label="Actual 2024-2026")
plt.plot(future_forecast.index, future_forecast, label="Forecast", linestyle="--", color="red")
plt.axvline(pd.Timestamp("2024-01-01"), color="black", linestyle=":", label="Forecast start")
plt.title("NHS A&E Attendances - 12 Month Forecast")
plt.xlabel("Date")
plt.ylabel("Monthly Attendances")
plt.legend()
plt.show()
```

```python
forecast_obj = results.get_forecast(steps=len(test) + future_steps)
forecast_mean = forecast_obj.predicted_mean
forecast_ci = forecast_obj.conf_int()
forecast_mean.index = future_index
forecast_ci.index = future_index

plt.figure(figsize=(14,5))
plt.plot(train.index, train, label="Historical data")
plt.plot(test.index, test, label="Actual 2024-2026")
plt.plot(forecast_mean.index, forecast_mean, label="Forecast", linestyle="--", color="red")
plt.fill_between(
    forecast_mean.index,
    forecast_ci.iloc[:, 0],
    forecast_ci.iloc[:, 1],
    alpha=0.2, color="red", label="95% confidence interval"
)
plt.axvline(pd.Timestamp("2024-01-01"), color="black", linestyle=":", label="Forecast start")
plt.title("NHS A&E Attendances - Forecast with Confidence Intervals")
plt.xlabel("Date")
plt.ylabel("Monthly Attendances")
plt.legend()
plt.show()
```
