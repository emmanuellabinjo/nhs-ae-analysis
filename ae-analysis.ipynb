```python
df = df[["Period", "Total Attendances"]].copy()
df.columns = ["date", "attendances"]
df = df.dropna()
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values("date")
print(df.shape)
print(df.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,5))
plt.plot(df["date"], df["attendances"])
plt.title("NHS A&E Total Attendances - England (2010-2026)")
plt.xlabel("Date")
plt.ylabel("Monthly Attendances")
plt.grid(True)
plt.show()
```

```python
df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month

import seaborn as sns

plt.figure(figsize=(12,5))
sns.boxplot(data=df[df["year"] < 2020], x="month", y="attendances")
plt.title("Seasonal pattern in A&E attendances (pre-COVID, 2010-2019)")
plt.xlabel("Month")
plt.ylabel("Monthly Attendances")
plt.show()
```

```python
from prophet import Prophet

prophet_df = df[["date", "attendances"]].copy()
prophet_df.columns = ["ds", "y"]

print(prophet_df.head())
```

```python
covid_lockdown = pd.DataFrame({
    "holiday": "covid_lockdown",
    "ds": pd.date_range(start="2020-03-01", end="2021-06-01", freq="MS"),
    "lower_window": 0,
    "upper_window": 0
})

print(covid_lockdown)
```

```python
model = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=False,
    daily_seasonality=False,
    holidays=covid_lockdown
)

model.fit(prophet_df)
```

```python
model = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=False,
    daily_seasonality=False,
    holidays=covid_lockdown,
    stan_backend="CMDSTANPY"
)

model.fit(prophet_df)
```
